```python
import pandas as pd
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
f1_data = pd.read_csv("/Users/kimseohee/Desktop/ELICE/F1-DataDriven/Data/circuits.csv")

# 'country' Ïó¥Î°ú Í∑∏Î£πÌôîÌïòÏó¨ ÏÑúÌÇ∑ Ïàò Í≥ÑÏÇ∞
country_counts = f1_data['country'].value_counts()

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ ÏÉâÏÉÅ ÏÑ§Ï†ï ('USA'ÏôÄ 'India'Îäî Îπ®Í∞ïÏÉâ, ÎÇòÎ®∏ÏßÄÎäî ÌååÎûëÏÉâ)
colors = ['red' if country in ['USA', 'India'] else 'blue' for country in country_counts.index]

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 3))
country_counts.plot(kind='bar', color=colors)

# Í∑∏ÎûòÌîÑ Ï†úÎ™©Í≥º Ï∂ï Î†àÏù¥Î∏î Ï∂îÍ∞Ä
plt.title('Number of Circuits by Country')
plt.xlabel('Country')
plt.ylabel('Number of Circuits')

plt.xticks(rotation=90)  # xÏ∂ï Î†àÏù¥Î∏îÏùò Í∞ÅÎèÑ Ï°∞Ï†à

# Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
plt.show()
```
